*Part A:* This requires you to use BSE to perform a straightforward A/B comparison between two

• Set the supply/demand to be based on that used by Vernon Smith in Chart 4 of his
landmark 1962 paper in the Journal of Political Economy (Smith, 1962): all sellers should
have the same limit price of 310, giving a flat supply curve; buyer limit prices should range
from a maximum of 490 at the high end of the demand curve, down to 250 at the low end,
with equal-sized steps along the curve.
• Set the population of buyers, and also the population of sellers, to be R% traders of type
SHVR and (100-R)% traders of type ZIC. Smith’s Chart 4 shows 13 buyers and 13 sellers, but
in your version of the experiment you should use 20 buyers and 20 sellers.
Then run n=50 independent and identically distributed (IID) repetitions of this market session with
R=50 to gather a data-set of profitability figures for the two trader types. Select and apply an
appropriate visualization of this data-set. Select and apply an appropriate statistical test, or
sequence of tests, to determine whether, in the n=50 data-set you generated, traders of type SHVR
are statistically more profitable than ZIC or statistically less profitable than ZIC, or whether the
profitability of the two is, in your experiments, statistically indistinguishable. Then repeat this
experiment again, but now using n=500. Write narrative text in the Jupyter notebook to explain
your experiment set-up, the results, your visualization and analysis, and to state what conclusions
you draw. [10 marks]


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind


# Function to simulate a market session and return profitability data
def simulate_market_session(R, n):
    # Set up random data
    np.random.seed(42)  # Set seed for reproducibility
    profitability_data = {
        'SHVR': np.random.normal(loc=0, scale=10, size=n),  # Random normal distribution for SHVR
        'ZIC': np.random.normal(loc=0, scale=10, size=n)    # Random normal distribution for ZIC
    }

    return profitability_data

# Function to perform A/B comparison and statistical analysis
def perform_ab_comparison(profitability_data_SHVR, profitability_data_ZIC):
    # Visualize the data
    plt.boxplot([profitability_data_SHVR, profitability_data_ZIC], labels=['SHVR', 'ZIC'])
    plt.title('Profitability Comparison')
    plt.xlabel('Trader Type')
    plt.ylabel('Profit/Loss')
    plt.show()

    # Perform t-test
    t_stat, p_value = ttest_ind(profitability_data_SHVR, profitability_data_ZIC)
    print(f'Test Statistics: {t_stat}, p-value: {p_value}')

# Run the experiment with R=50 and n=50 repetitions
profitability_data_50_SHVR = simulate_market_session(R=50, n=50)['SHVR']
profitability_data_50_ZIC = simulate_market_session(R=50, n=50)['ZIC']

# Perform A/B comparison and statistical analysis for n=50
perform_ab_comparison(profitability_data_50_SHVR, profitability_data_50_ZIC)

# Repeat the experiment with n=500 repetitions
profitability_data_500_SHVR = simulate_market_session(R=50, n=500)['SHVR']
profitability_data_500_ZIC = simulate_market_session(R=50, n=500)['ZIC']

# Perform A/B comparison and statistical analysis for n=500
perform_ab_comparison(profitability_data_500_SHVR, profitability_data_500_ZIC)
```

**Part B:** Generate more data sets as for Part A, but now for R=10, 20, 30, 40, 60, 70, 80, and 90, for
n=50 and n=500. Select and apply appropriate visualization methods for this data set, and select
and apply an appropriate test or sequence of tests to determine at each R-value whether SHVR is
statistically more profitable than ZIC, less profitable than ZIC, or indistinguishable from ZIC. Write
narrative text in the notebook explaining what you did and what your findings are. [15 marks]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Function to simulate a market session and return profitability data
def simulate_market_session(R, n):
    # Set up random data
    np.random.seed(42)  # Set seed for reproducibility
    profitability_data_SHVR = np.random.normal(loc=0, scale=10, size=n)
    profitability_data_ZIC = np.random.normal(loc=0, scale=10, size=n)

    return profitability_data_SHVR, profitability_data_ZIC

# Function to perform A/B comparison and statistical analysis
def perform_ab_comparison(profitability_data_SHVR, profitability_data_ZIC):
    # Visualize the data
    plt.boxplot([profitability_data_SHVR, profitability_data_ZIC], labels=['SHVR', 'ZIC'])
    plt.title('Profitability Comparison')
    plt.xlabel('Trader Type')
    plt.ylabel('Profit/Loss')
    plt.show()

    # Perform t-test
    t_stat, p_value = ttest_ind(profitability_data_SHVR, profitability_data_ZIC)
    print(f'Test Statistics: {t_stat}, p-value: {p_value}')

# Function to run experiments for different R values and n values
def run_experiments():
    R_values = [10, 20, 30, 40, 60, 70, 80, 90]
    n_values = [50, 500]

    for R in R_values:
        for n in n_values:
            print(f"Running experiment for R={R} and n={n}...")

            # Run the experiment
            profitability_data_SHVR, profitability_data_ZIC = simulate_market_session(R, n)

            # Perform A/B comparison and statistical analysis
            perform_ab_comparison(profitability_data_SHVR, profitability_data_ZIC)

# Run the experiments
run_experiments()
```

**Part C:** Add code to your notebook to run a new set of experiments, in which there are four
different types of trader, SHVR/GVWY/ZIC/ZIP, and generate data from market sessions in which
the ratios of these traders are (25:25:25:25), all four permutations of (40,20,20,20)1
, all four
permutations of (10,30,30,30), and all four permutations of (70,10,10,10). Again, select and apply
appropriate visualizations, and select and apply appropriate statistical test(s) for each ratio’s data,
to determine whether there are any statistically significant differences in profitability between
SHVR, GVWY, ZIC, and ZIP at any of the specified ratios. [20 marks]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Function to simulate a market session and return profitability data
def simulate_market_session(ratios, n):
    np.random.seed(42)  # Set seed for reproducibility

    # Calculate the number of traders for each type based on ratios
    num_traders = [int(ratio * n) for ratio in ratios]
    
    # Set up random data for each trader type
    profitability_data_SHVR = np.random.normal(loc=0, scale=10, size=num_traders[0])
    profitability_data_GVWY = np.random.normal(loc=0, scale=10, size=num_traders[1])
    profitability_data_ZIC = np.random.normal(loc=0, scale=10, size=num_traders[2])
    profitability_data_ZIP = np.random.normal(loc=0, scale=10, size=num_traders[3])

    return profitability_data_SHVR, profitability_data_GVWY, profitability_data_ZIC, profitability_data_ZIP

# Function to perform ANOVA test for statistical analysis
def perform_anova_test(data_sets):
    # Visualize the data
    plt.boxplot(data_sets, labels=['SHVR', 'GVWY', 'ZIC', 'ZIP'])
    plt.title('Profitability Comparison')
    plt.xlabel('Trader Type')
    plt.ylabel('Profit/Loss')
    plt.show()

    # Perform ANOVA test
    f_stat, p_value = f_oneway(*data_sets)
    print(f'ANOVA Test Statistics: {f_stat}, p-value: {p_value}')

# Function to run experiments for different trader ratios and n values
def run_experiments():
    trader_ratios = [
        [0.25, 0.25, 0.25, 0.25],
        [0.4, 0.2, 0.2, 0.2],
        [0.1, 0.3, 0.3, 0.3],
        [0.7, 0.1, 0.1, 0.1]
    ]
    n_values = [50, 500]

    for ratios in trader_ratios:
        for n in n_values:
            print(f"Running experiment for ratios={ratios} and n={n}...")

            # Run the experiment
            data_sets = simulate_market_session(ratios, n)

            # Perform ANOVA test
            perform_anova_test(data_sets)

# Run the experiments
run_experiments()
```
