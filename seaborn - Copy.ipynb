```python
import pandas as pd
import numpy as np

# Creating a dummy dataset
np.random.seed(42)
data = {
    'Category': np.random.choice(['A', 'B', 'C', 'D'], size=100),
    'Value1': np.random.normal(0, 1, 100),
    'Value2': np.random.normal(10, 2, 100),
    'Day': np.random.choice(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'], size=100),
    'Gender': np.random.choice(['Male', 'Female'], size=100)
}

df = pd.DataFrame(data)
df
```

```python
import seaborn as sns
```

```python
sns.set() # set default themea
```

```python
#scatter plots
sns.scatterplot(x = 'Value1', y = 'Value2', data = df)
```

```python
#lienplot
sns.lineplot(x = 'Value1', y = 'Value2', data = df)
```

```python
#BARPLOT
sns.barplot(x = 'Value1', y = 'Value2', data = df)
```

```python
#Boxplot
sns.boxplot(x = 'Value1', y = 'Value2', data = df)
```

```python
#distribution plot
sns.histplot(x = 'Value1', y = 'Value2', data = df)
```

```python
#kernel ldenisty plot
sns.kdeplot(x = 'Value1',  data = df)
```

```python
#kde with 2 sides
sns.kdeplot(x = 'Value1', y = 'Value2', data = df)
```

```python
#kde with 2 sides
#kernel ldenisty plot
sns.kdeplot(x = 'Value1',  data = df)

sns.kdeplot(x = 'Value1', y = 'Value2', data = df)
```

```python
#pair plots
sns.pairplot(df[['Value1', 'Value2','Category']], hue = 'Category')
```

```python
#heatmaps
corelation_matrix = df[['Value1', 'Value2']].corr()
sns.heatmap(corelation_matrix, annot = True, cmap = 'coolwarm')
```

```python
#customissing theme
```

```python
#wee have been using default theme now we will use  pernsonlized theme
sns.set_palette('husl')
sns.set_style('whitegrid')
```

```python
#setting title and labels
#for plt to work import matplotlib as plt
import matplotlib.pyplot as plt
#note thsi will result in errors as we dont have a plot with function plt
plt.title('Tittle')
plt.xlabel('X-alebl')
plt.ylabel('y-alebl')
```

```python
#avanced grids
g = sns.FacetGrid(df, col = 'Day', hue = 'Gender')
#lets fix this by adding more features
g.map(plt.scatter, 'Value1', 'Value2', alpha = 0.7)
g.add_legend()
```

```python
#regression line tih hue
sns.lmplot(x = 'Value1', y = 'Value2', data = df, hue = 'Gender')
```

```python
g = sns.PairGrid(df, hue = 'Category')
g.map_diag(plt.hist)
g.add_legend()
```

```python
#custome pa;ettes
custom_palette = sns.color_palette("husl",4)
sns.boxplot(x = 'Category', y = 'Value1', data = df, palette = custom_palette)
```
