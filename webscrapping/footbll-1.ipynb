```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # We'll use seaborn for color mapping

# Assuming combined_df is your DataFrame with "Player," "MP," and "Nation" columns

# Group the DataFrame by the "Player" column and calculate the sum of "MP" for each player
player_mp_sum = combined_df.groupby('Player')['MP'].sum()

# Sort the player_mp_sum Series in descending order
player_mp_sum = player_mp_sum.sort_values(ascending=False)


# Get unique colors for each unique "Nation"
unique_nations = combined_df['Nation'].unique()
colors = sns.color_palette('Set3', len(unique_nations))

# Create a color dictionary for mapping "Nation" to color
color_dict = {nation: color for nation, color in zip(unique_nations, colors)}

# Map colors based on "Nation" for each player
player_colors = combined_df['Nation'].map(color_dict)

# Create a bar plot for the sum of "MP" by player with colors
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
player_mp_sum.plot(kind='bar', color=player_colors)  # Create the bar plot with mapped colors

plt.xlabel('Player')  # X-axis label
plt.ylabel('Total Matches Played (MP)')   # Y-axis label
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility (optional)
plt.title('Total Matches Played (MP) by Player (Sorted)')  # Title of the plot

# Create a legend based on unique "Nation" values and their corresponding colors
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=nation, markersize=10, markerfacecolor=color)
                   for nation, color in color_dict.items()]

# Place the legend outside the plot to the right
plt.legend(handles=legend_elements, title='Nation', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()  # Ensures that labels are not cut off
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming combined_df is your DataFrame with "Player," "Goals," and "Nation" columns

# Group the DataFrame by the "Player" column and calculate the sum of "Goals" for each player
player_goals_sum = combined_df.groupby('Player')['Goals'].sum()

# Sort the player_goals_sum Series in descending order
player_goals_sum = player_goals_sum.sort_values(ascending=False)

# Get unique colors for each unique "Nation" using seaborn
unique_nations = combined_df['Nation'].unique()
colors = sns.color_palette('Set3', len(unique_nations))

# Create a color dictionary for mapping "Nation" to color
color_dict = {nation: color for nation, color in zip(unique_nations, colors)}

# Map colors based on "Nation" for each player
player_colors = combined_df['Nation'].map(color_dict)

# Create a bar plot for the sum of "Goals" by player with colors
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
player_goals_sum.plot(kind='bar', color=player_colors)  # Create the bar plot with mapped colors

plt.xlabel('Player')  # X-axis label
plt.ylabel('Total Goals')   # Y-axis label
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility (optional)
plt.title('Total Goals by Player (Sorted)')  # Title of the plot

# Create a legend based on unique "Nation" values and their corresponding colors
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=nation, markersize=10, markerfacecolor=color)
                   for nation, color in color_dict.items()]

# Place the legend outside the plot to the right
plt.legend(handles=legend_elements, title='Nation', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()  # Ensures that labels are not cut off
plt.show()
```
