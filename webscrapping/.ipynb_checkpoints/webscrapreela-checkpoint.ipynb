```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Create a new instance of the Firefox web driver
driver = webdriver.Firefox()

# Create lists to store titles, links, and news
titles = []
links = []
news = []

# Specify the number of pages to iterate through
num_pages = 10

for page in range(1, num_pages + 1):
    # Construct the URL for each page
    url = f"https://www.kathimerini.gr/epikairothta/page/{page}/"
    driver.get(url)

    # Wait for the page to load completely (you can adjust the sleep time as needed)
    time.sleep(5)

    # Find all <a> elements with class "py-4 mainlink" by CSS selector
    link_elements = driver.find_elements(By.CSS_SELECTOR, "a.py-4.mainlink")

    # Extract the href attributes from all link elements and store in a list
    links.extend([link.get_attribute("href") for link in link_elements])

    # Find all <span> elements with class "card-title" by CSS selector
    title_elements = driver.find_elements(By.CSS_SELECTOR, "span.card-title")

    # Extract the text from all title elements and store in the 'titles' list
    titles.extend([title.text for title in title_elements])

    # Iterate through the collected links and open each in a new tab
    for link in links:
        driver.execute_script("window.open('" + link + "', '_blank');")
        driver.switch_to.window(driver.window_handles[-1])

        # Wait for the new page to load (you can adjust the sleep time as needed)
        time.sleep(5)

        # Find all <p> elements without any class by CSS selector
        p_elements = driver.find_elements(By.CSS_SELECTOR, "p:not([class])")

        # Extract the text from all p elements and store in the 'news' list
        news_text = [p.text for p in p_elements]
        news.append(' '.join(news_text))

        # Close the current tab
        driver.close()

        # Switch back to the original tab
        driver.switch_to.window(driver.window_handles[0])

# Ensure all lists are of the same length
min_length = min(len(titles), len(links), len(news))
titles = titles[:min_length]
links = links[:min_length]
news = news[:min_length]

# Create a DataFrame with the extracted titles, links, and news
data = {'Title': titles, 'Link': links, 'News': news}
df1 = pd.DataFrame(data)

# Print the DataFrame
print(df1)

# Close the web driver when done
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Create options for a headless browser
options = webdriver.FirefoxOptions()
options.headless = True

# Create a new instance of the Firefox web driver with headless options
driver = webdriver.Firefox(options=options)

# Open the specified link
url = "https://www.ethnos.gr/news"
driver.get(url)

# Wait for the page to load completely (you can adjust the sleep time as needed)
driver.implicitly_wait(5)

# Initialize a variable to keep track of the collected entries
collected_entries = 0

# Create empty lists to store the data
dateposted = []
single_titles = []
news_text = []

# Set the target number of entries
target_entries = 10 # Change this to the desired number of entries

# Loop until the target number of entries is reached
while collected_entries < target_entries:
    # Find all <div> elements with class "single-category" by CSS selector
    div_elements = driver.find_elements(By.CSS_SELECTOR, "div.single-category")

    # Find all <a> elements with class "single-title" by CSS selector
    title_elements = driver.find_elements(By.CSS_SELECTOR, "a.single-title")

    # Extract the text from all div and title elements and store in the respective lists
    dateposted.extend([div.text for div in div_elements])
    single_titles.extend([title.text for title in title_elements])

    # Update the count of collected entries
    collected_entries = len(dateposted)

    # Scroll down to trigger more content loading (you may need to adjust the scroll amount)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait for a short time to allow potential dynamic content loading
    driver.implicitly_wait(3)

# Iterate through the collected links and open each in a new tab
for link in single_titles:
    driver.execute_script("window.open('" + link + "', '_blank');")
    driver.switch_to.window(driver.window_handles[-1])

    # Find all empty <p> elements within the div with class "content-section"
    p_elements = driver.find_elements(By.CSS_SELECTOR, "div.content-section p:empty")

    # Extract the text from all empty p elements and store in the 'news_text' list
    news_text.append(' '.join([p.text for p in p_elements]))

    # Close the current tab
    driver.close()

    # Switch back to the original tab
    driver.switch_to.window(driver.window_handles[0])

# Trim the lists to the target number of entries
dateposted = dateposted[:target_entries]
single_titles = single_titles[:target_entries]
news_text = news_text[:target_entries]

# Create a DataFrame with the extracted data
df3 = pd.DataFrame({'DatePosted': dateposted, 'SingleTitle': single_titles})#, 'NewsText': news_text})

# Print the DataFrame
print(df3)

# Close the web driver when done
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Create a new instance of the Firefox web driver
driver = webdriver.Firefox()

# Set the number of pages to loop through (in this example, up to 5 pages)
num_pages = 10

# Initialize empty lists to store data
all_titles = []
all_links = []
all_time_posted = []
all_news = []

for page in range(1, num_pages + 1):
    # Construct the URL for each page
    url = f"https://thepressproject.gr/article_type/newsroom/page/{page}/"
    driver.get(url)

    # Optional: Wait for the page to load completely (you can adjust the sleep time as needed)
    driver.implicitly_wait(5)

    # Find all h3 elements on the page
    h3_elements = driver.find_elements(By.TAG_NAME, "h3")

    # Extract text from h3 and store in the 'all_titles' list
    titles = [h3.text for h3 in h3_elements]

    # Find all div elements with class "col-md-8 archive-item"
    div_elements = driver.find_elements(By.CSS_SELECTOR, "div.col-md-8.archive-item")

    # Extract links, timeposted, and news from anchor tags and div elements
    for div in div_elements:
        link = div.find_element(By.TAG_NAME, "a").get_attribute("href")
        time_posted = div.find_element(By.CSS_SELECTOR, "div.archive-info.info-text").text
        news = div.find_element(By.CSS_SELECTOR, "div.archive-body").text

        all_links.append(link)
        all_titles.append(titles.pop(0))  # Pop the corresponding title for each link
        all_time_posted.append(time_posted)
        all_news.append(news)

# Create a DataFrame with the extracted data
df3 = pd.DataFrame({
    "titles": all_titles,
    "link": all_links,
    "time_Posted": all_time_posted,
    "news": all_news
})

# Print the DataFrame
print(df3)

# Close the web driver when done
driver.quit()
```

```python
# 4
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Create a new instance of the Firefox web driver
driver = webdriver.Firefox()

# Open the specified link
url = "https://startupper.gr/news/"
driver.get(url)

# Optional: Wait for the page to load completely (you can adjust the sleep time as needed)
driver.implicitly_wait(5)

# Initialize empty lists to store titles and news
all_titles = []
all_news = []

# Specify the number of pages to iterate through
num_pages = 10

for page in range(num_pages):
    # Find all h2 elements with class "title"
    h2_elements = driver.find_elements(By.CSS_SELECTOR, "h2.title")
    
    # Extract text from h2 elements and store in the 'all_titles' list
    all_titles.extend([h2.text for h2 in h2_elements])

    # Find all div elements with class "post-summary"
    summary_elements = driver.find_elements(By.CSS_SELECTOR, "div.post-summary")
    
    # Extract text from div elements and store in the 'all_news' list
    all_news.extend([summary.text for summary in summary_elements])

    # Scroll to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Wait for a short time to allow content to load (you can adjust the sleep time as needed)
    time.sleep(2)

# Ensure both lists are of the same length
min_length = min(len(all_titles), len(all_news))
all_titles = all_titles[:min_length]
all_news = all_news[:min_length]

# Create a DataFrame with the extracted titles and news
df4 = pd.DataFrame({"Title": all_titles, "News": all_news})

# Print the DataFrame
print(df4)

# Close the web driver when done
driver.quit()
```

```python
#5
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Create a new instance of the Firefox web driver
driver = webdriver.Firefox()

# Open the specified link
url = "https://www.naftemporiki.gr/english/"
driver.get(url)

# Optional: Wait for the page to load completely (you can adjust the sleep time as needed)
driver.implicitly_wait(5)

# Initialize empty lists to store titles, links, and news
titles = []
links = []
news = []

# Loop through multiple pages (adjust the range as needed)
for page in range(1, 10):  # Example: Loop through 5 pages
    # Find all div elements with class "g-col-lg-9 g-col-8"
    div_elements = driver.find_elements(By.CSS_SELECTOR, "div.g-col-lg-9.g-col-8")

    # Loop through each div element and extract titles, links, and news
    for div_element in div_elements:
        # Find all anchor elements within the div
        anchor_elements = div_element.find_elements(By.TAG_NAME, "a")

        # Extract text from h2 elements and store in the 'titles' list
        titles.extend([anchor.text for anchor in anchor_elements])

        # Extract href attributes from anchor elements and store in the 'links' list
        links.extend([anchor.get_attribute("href") for anchor in anchor_elements])

        # Iterate through the collected links and open each in a new tab
        for link in links:
            driver.execute_script("window.open('" + link + "', '_blank');")
            driver.switch_to.window(driver.window_handles[-1])

            # Wait for the new page to load (you can adjust the sleep time as needed)
            driver.implicitly_wait(5)

            # Find the div element with class "post-content mb-25" and extract the text
            try:
                news_element = driver.find_element(By.CSS_SELECTOR, "div.post-content.mb-25")
                news_text = news_element.text
                news.append(news_text)
            except:
                news.append("")  # If the element is not found, append an empty string

            # Close the current tab
            driver.close()

            # Switch back to the original tab
            driver.switch_to.window(driver.window_handles[0])

    # Navigate to the next page (if applicable)
    try:
        next_page_button = driver.find_element(By.XPATH, "//a[@rel='next']")
        next_page_button.click()
        # Optional: Wait for the page to load completely after clicking the next page button
        driver.implicitly_wait(5)
    except:
        break  # Break the loop if there is no next page button

# Ensure all lists are of the same length
min_length = min(len(titles), len(links), len(news))
titles = titles[:min_length]
links = links[:min_length]
news = news[:min_length]

# Create a DataFrame with the extracted titles, links, and news
df5 = pd.DataFrame({"Title": titles, "Link": links, "News": news})

# Print the DataFrame
print(df5)

# Close the web driver when done
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Create a new instance of the Firefox web driver
driver = webdriver.Firefox()

# Open the specified link
url = "https://www.capital.gr/eidiseis/"
driver.get(url)

# Optional: Wait for the page to load completely (you can adjust the sleep time as needed)
driver.implicitly_wait(5)

# Initialize empty lists to store titles, links, and news
titles = []
links = []
news = []

# Define the number of pages to scrape
num_pages = 10  # You can adjust this number as needed

# Loop through multiple pages
for _ in range(num_pages):
    # Find all h2 elements with class "bold" that have an anchor tag
    h2_elements = driver.find_elements(By.CSS_SELECTOR, "h2.bold a")

    # Extract text from h2 elements and store in the 'titles' list
    titles.extend([h2.text for h2 in h2_elements])

    # Extract the links from the anchor tags within each h2 element
    links.extend([h2.get_attribute("href") for h2 in h2_elements])

    # Navigate to the next page (if applicable)
    try:
        next_page_button = driver.find_element(By.XPATH, "//a[@class='pagination-next']")
        next_page_button.click()
        # Optional: Wait for the page to load completely after clicking the next page button
        driver.implicitly_wait(5)
    except:
        break  # Break the loop if there is no next page button

# Open each link collected and extract information from the div tag with id "articleBody"
for link in links:
    if link:
        # Open the link
        driver.get(link)

        # Find the div tag with id "articleBody" and extract the text
        try:
            news_text = driver.find_element(By.ID, "articleBody").text
            news.append(news_text)
        except:
            news.append("No information found")

# Create a DataFrame with the extracted titles, links, and news
df6 = pd.DataFrame({"Title": titles, "Link": links, "News": news})

# Print the DataFrame
print(df6)

# Close the web driver when done
driver.quit()
```

```python

# Define a function to merge columns with similar names
def merge_columns_with_keyword(df, keyword):
    relevant_columns = [col for col in df.columns if keyword.lower() in col.lower()]
    merged_column = df[relevant_columns].apply(lambda row: ' '.join(row.dropna().astype(str)), axis=1)
    return merged_column.rename(f"{keyword.capitalize()}")

# Merge columns with "news" in the name
news_column = merge_columns_with_keyword(pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True), "news")

# Merge columns with "title" in the name
title_column = merge_columns_with_keyword(pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True), "title")

# Concatenate the resulting columns horizontally
final_df = pd.concat([title_column, news_column], axis=1)

final_df
```
